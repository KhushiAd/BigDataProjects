```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\khush\Documents\BigData\Projects\Project3\Athena\results.csv")
df
```

```python
#Graph the maximum volatility trend per company (A single Line Chart: Each line refers to a company)
high_vol = df.drop(columns=["Average_Volatility","Lowest_Volatility"]).pivot(index="Date",columns="Company_Name",values="Highest_Volatility")
high_vol
tickers = high_vol.columns

for ticker in tickers:
    plt.plot(high_vol.index,high_vol[ticker],label=ticker)
plt.gcf().autofmt_xdate(rotation=25)
plt.legend(loc="right")
plt.show()
```

```python
#According to the plot, costco looks to be the most volatile.
```

```python
#Graph the daily average volatility per company(A Grouped Bar Chart: Each group refers to a company and the bars refer to the daily highest volatility)
avg_vol = df.groupby('Company_Name')["Average_Volatility"].mean()
avg_vol.plot.bar()
```

```python
# Lowest Daily Maximum Volatility oer company
highest_vol = df.groupby('Company_Name')["Highest_Volatility"].agg('max')
highest_vol.plot.bar()
```
